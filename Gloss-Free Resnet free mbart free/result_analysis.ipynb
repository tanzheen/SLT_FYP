```python
# Path to the text file
from sacrebleu import BLEU
file_path1 = "dev_ref.txt"
# Initialize an empty list
dev_ref = []
# Open the file and read each line into the list
with open(file_path1, "r", encoding="utf-8") as file:
    dev_ref = [line.strip() for line in file]  # Use strip() to remove leading/trailing whitespace

# Path to the text file
file_path2 = "dev_pred.txt"
# Initialize an empty list
dev_pred = []
# Open the file and read each line into the list
with open(file_path2, "r", encoding="utf-8") as file:
    dev_pred = [line.strip() for line in file]  # Use strip() to remove leading/trailing whitespace

assert len(dev_ref) == len(dev_pred), "The number of lines in the two files do not match"
print("Metrics for the dev set:")
for i in range(1,5): 
    # Initialize BLEU scorer
    bleu = BLEU(max_ngram_order=i)

    # Calculate corpus BLEU score
    bleu_s = bleu.corpus_score(dev_pred, [dev_ref]).score
    print(f"Corpus BLEU-{i} score: {bleu_s:.2f}")
    
```
