## Processing the weights for transfer

```python
import torch

# Load visual encoder weights
visual_encoder_weights = torch.load("unwrapped_model/pytorch_model.bin", map_location="cpu")

# Load text decoder weights
text_decoder_weights = torch.load("text_decoder/pytorch_model.bin", map_location="cpu")
```

```python
print("Visual Encoder Layer Names:")
for key in visual_encoder_weights.keys():
    print(key)

```

```python
print("\nText Decoder Layer Names:")
for key in text_decoder_weights.keys():
    print(key)
```

```python
from transformers import MBartForConditionalGeneration

# Load a pre-trained MBart model
mbart_model = MBartForConditionalGeneration.from_pretrained("pretrain_models/MBart_proun")
```

```python
# Create a new state dict for MBart decoder
mbart_decoder_weights = {}

# Assuming the text decoder keys align with MBart's decoder keys (check naming carefully)
for key, value in text_decoder_weights.items():
    # Replace text_decoder's layer names to match MBart's decoder layers
    new_key = key.replace("text_decoder", "model.decoder")
    mbart_decoder_weights[new_key] = value
    print(new_key)
```

```python
# Create a new state dict for MBart encoder
mbart_encoder_weights = {}

# Assuming the visual encoder keys align with MBart's encoder keys (check naming carefully)
for key, value in visual_encoder_weights.items():
    # Replace visual_encoder's layer names to match MBart's encoder layers
    new_key = key.replace("model_images.trans_encoder", "model.encoder")
    mbart_encoder_weights[new_key] = value
    print(new_key)

# Initialize an empty dictionary to store only the relevant encoder weights
filtered_mbart_encoder_weights = {}

# Loop through the mapped keys and filter those that start with "model.encoder"
for key, value in mbart_encoder_weights.items():
    if key.startswith("model.encoder"):
        filtered_mbart_encoder_weights[key] = value

# Print to verify the filtered keys
for key in filtered_mbart_encoder_weights:
    print(key)
```

```python
# Get the full MBart state dict
mbart_state_dict = mbart_model.state_dict()

# Update MBart's encoder and decoder with the respective weights
mbart_state_dict.update(filtered_mbart_encoder_weights )
mbart_state_dict.update(mbart_decoder_weights)
```

```python
# Load the merged state dict into the MBart model
mbart_model.load_state_dict(mbart_state_dict, strict=False)
```

```python
from transformers import MBartForConditionalGeneration, MBartTokenizer
import torch

# Initialize the model and tokenizer
model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-cc25")
tokenizer = MBartTokenizer.from_pretrained("facebook/mbart-large-cc25")

# Sample input sentence
sentence = ""

# Encode the input sentence
encoded_input = tokenizer(sentence, return_tensors="pt")

# Forward pass through the encoder only to get encoder embeddings
with torch.no_grad():
    encoder_outputs = model.model.encoder(
        input_ids=encoded_input['input_ids'],
        attention_mask=encoded_input['attention_mask'],
        return_dict=True
    )

encoder_embeddings = encoder_outputs.last_hidden_state
print("Encoder Embeddings Shape:", encoder_embeddings.shape)
print("Encoder Embeddings:", encoder_embeddings)
```

```python
from transformers import MBartForConditionalGeneration, MBart50Tokenizer

# Load MBart model and tokenizer for Chinese-English translation
model_name = "facebook/mbart-large-50-many-to-many-mmt"
model = MBartForConditionalGeneration.from_pretrained(model_name)
tokenizer = MBart50Tokenizer.from_pretrained(model_name)
tokenizer.src_lang = "zh_CN"  # Set source language to Simplified Chinese

# Input Chinese sentence
sentence = ""

# Encode the sentence
encoded_input = tokenizer(sentence, return_tensors="pt")

# Generate translation (you can adjust num_beams and other generation parameters as needed)
output_tokens = model.generate(**encoded_input, forced_bos_token_id=tokenizer.lang_code_to_id["zh_CN"])

# Decode the generated tokens to get the translated sentence
output_sentence = tokenizer.decode(output_tokens[0], skip_special_tokens=True)

print("Input Sentence:", sentence)
print("Translated Output:", output_sentence)
```

```python
import torch
from transformers import MBartForConditionalGeneration, MBart50Tokenizer
import torch.nn as nn

# Load the MBart model and tokenizer for Chinese-to-English translation
model_name = "facebook/mbart-large-50-many-to-many-mmt"
model = MBartForConditionalGeneration.from_pretrained(model_name)
tokenizer = MBart50Tokenizer.from_pretrained(model_name)
tokenizer.src_lang = "zh_CN"  # Set source language to Simplified Chinese

# Input and target sentences
input_sentence = ""  # Source sentence in Chinese
target_sentence = "Hello! This is a test."  # Target sentence in English

# Tokenize the input and target sentences
input_tokens = tokenizer(input_sentence, return_tensors="pt")
target_tokens = tokenizer(target_sentence, return_tensors="pt")

# Pass input_ids and labels (target token ids) to the model
# MBart will automatically calculate the cross-entropy loss
outputs = model(input_ids=input_tokens['input_ids'], labels=target_tokens['input_ids'])

# The seq2seq loss is in outputs.loss
seq2seq_loss = outputs.loss
print("Seq2Seq Loss:", seq2seq_loss.item())


loss_fct = nn.CrossEntropyLoss()
masked_lm_loss = loss_fct(outputs.logits.view(-1, outputs.logits.shape[-1]), target_tokens['input_ids'].view(-1)) #* args.loss_lambda
print("Masked LM Loss:", masked_lm_loss.item())
```
